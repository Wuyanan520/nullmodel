```python
call = pd.read_csv('../tb_call_201202.txt',sep='\t',header=None,usecols=[1,11])
call.columns = ['calling_nbr','raw_dur']
call.head()
```

```python
user_info = pd.read_csv('../tb_user_info_201202.txt',sep='\t',header=None,usecols=[0,3])
user_info.columns = ['ACC_NBR','GENDER']
user_info.head()
```

```python
# 查看男女分布
user_info.GENDER.value_counts()
```

---

# 分析

```python
# 用户的呼叫时长累积分布
call_time = call.groupby('calling_nbr')['raw_dur'].sum().sort_values()
call_time = pd.DataFrame(call_time)
call_time.columns = ['durtime']
call_time
```

```python
# 用户的呼叫次数分布
call_num = call.groupby('calling_nbr')['raw_dur'].count().sort_values()
call_num = pd.DataFrame(call_num)
call_num.columns = ['num']
call_num
```

```python
call_feature = call_time.join(call_num)
call_feature = call_feature.join(user_info.set_index('ACC_NBR'))
call_feature
```

```python
call_feature.GENDER.value_counts()
```

```python
print "男性平均每月打电话 ", call_feature[call_feature.GENDER=='M'].num.mean(), '次'
print "女性平均每月打电话 ", call_feature[call_feature.GENDER=='F'].num.mean(), '次'
```

```python
print "男性平均每月打电话持续时间 ", call_feature[call_feature.GENDER=='M'].durtime.mean(), '秒'
print "女性平均每月打电话持续时间 ", call_feature[call_feature.GENDER=='F'].durtime.mean(), '秒'
```

```python
call_feature.to_csv('call_feature.csv')
```
